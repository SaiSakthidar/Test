```python
# GOOGLE COLAB AUDIO RECORDING SYSTEM
# This cell provides browser-based audio recording for Google Colab users

import sys
import base64
import io
from IPython.display import HTML, display, Javascript, Audio as IPythonAudio
from google.colab import output
import json
import time

def create_colab_audio_recorder():
    """Create a JavaScript-based audio recorder for Google Colab"""

    # Check if we're in Colab
    if 'google.colab' not in sys.modules:
        print("⚠️ This cell is designed for Google Colab.")
        print("For local usage, use the continuous conversation system instead.")
        return

    print("🎙️ Initializing Google Colab Audio Recorder...")

    # JavaScript code for audio recording in Colab
    js_code = """
    async function recordAudio() {
        const recordButton = document.getElementById('recordButton');
        const statusDiv = document.getElementById('status');
        const audioPlayer = document.getElementById('audioPlayer');

        if (!navigator.mediaDevices || !navigator.mediaDevices.getUserMedia) {
            statusDiv.innerHTML = '❌ Audio recording not supported in this browser';
            return;
        }

        let mediaRecorder;
        let audioChunks = [];
        let isRecording = false;

        try {
            const stream = await navigator.mediaDevices.getUserMedia({
                audio: {
                    sampleRate: 16000,
                    channelCount: 1,
                    echoCancellation: true,
                    noiseSuppression: true
                }
            });

            mediaRecorder = new MediaRecorder(stream, {
                mimeType: 'audio/webm;codecs=opus'
            });

            mediaRecorder.ondataavailable = event => {
                if (event.data.size > 0) {
                    audioChunks.push(event.data);
                }
            };

            mediaRecorder.onstop = async () => {
                const audioBlob = new Blob(audioChunks, { type: 'audio/webm' });
                const audioUrl = URL.createObjectURL(audioBlob);
                audioPlayer.src = audioUrl;
                audioPlayer.style.display = 'block';

                // Convert to base64 for Python
                const reader = new FileReader();
                reader.onloadend = () => {
                    const base64Audio = reader.result.split(',')[1];
                    statusDiv.innerHTML = '✅ Recording complete! Processing...';

                    // Send to Python
                    google.colab.kernel.invokeFunction('process_colab_audio', [base64Audio], {});
                };
                reader.readAsDataURL(audioBlob);

                // Stop all tracks
                stream.getTracks().forEach(track => track.stop());
            };

            // Recording button click handler
            recordButton.onclick = () => {
                if (!isRecording) {
                    audioChunks = [];
                    mediaRecorder.start();
                    isRecording = true;
                    recordButton.innerHTML = '⏹️ Stop Recording (Max 10s)';
                    recordButton.style.backgroundColor = '#ff4444';
                    statusDiv.innerHTML = '🔴 Recording... Speak now!';

                    // Auto-stop after 10 seconds
                    setTimeout(() => {
                        if (isRecording) {
                            mediaRecorder.stop();
                            isRecording = false;
                            recordButton.innerHTML = '🎤 Start Recording';
                            recordButton.style.backgroundColor = '#4CAF50';
                        }
                    }, 10000);
                } else {
                    mediaRecorder.stop();
                    isRecording = false;
                    recordButton.innerHTML = '🎤 Start Recording';
                    recordButton.style.backgroundColor = '#4CAF50';
                }
            };

            recordButton.disabled = false;
            recordButton.innerHTML = '🎤 Start Recording';
            statusDiv.innerHTML = '✅ Microphone ready! Click to start recording.';

        } catch (error) {
            console.error('Error accessing microphone:', error);
            statusDiv.innerHTML = '❌ Could not access microphone. ' + error.message;
            recordButton.innerHTML = '❌ Mic Access Denied';
            recordButton.disabled = true;
        }
    }

    // Initialize when DOM is ready
    if (document.readyState === 'loading') {
        document.addEventListener('DOMContentLoaded', recordAudio);
    } else {
        recordAudio();
    }
    """

    # HTML interface
    html_interface = f"""
    <div style="border: 2px solid #4CAF50; border-radius: 10px; padding: 20px; margin: 10px 0; background-color: #f9f9f9;">
        <h3 style="color: #4CAF50;">🎙️ Google Colab Audio Recorder</h3>
        <p><strong>Instructions:</strong></p>
        <ol>
            <li>Click "Start Recording" and allow microphone access when prompted</li>
            <li>Speak clearly for up to 10 seconds</li>
            <li>Click "Stop Recording" or wait for auto-stop</li>
            <li>The audio will be processed by Gemma 3N automatically</li>
        </ol>
        <button id="recordButton" disabled style="
            background-color: #cccccc;
            color: white;
            border: none;
            padding: 15px 30px;
            font-size: 16px;
            border-radius: 5px;
            cursor: pointer;
            margin: 10px 0;
        ">🎤 Initializing...</button>
        <div id="status" style="margin: 10px 0; font-weight: bold;">⏳ Setting up microphone...</div>
        <audio id="audioPlayer" controls style="display: none; width: 100%; margin: 10px 0;"></audio>
    </div>
    <script>
    {js_code}
    </script>
    """

    display(HTML(html_interface))

    return True

# Global variable to store the latest audio
latest_colab_audio = None

def process_colab_audio(audio_base64):
    """Process audio recorded in Colab"""
    global latest_colab_audio

    try:
        print("🎵 Processing Colab audio recording...")

        # Decode base64 audio
        audio_bytes = base64.b64decode(audio_base64)
        latest_colab_audio = audio_bytes

        print(f"✅ Audio received: {len(audio_bytes)} bytes")

        # Save to temporary file for processing
        import tempfile
        with tempfile.NamedTemporaryFile(suffix='.webm', delete=False) as tmp_file:
            tmp_file.write(audio_bytes)
            temp_audio_path = tmp_file.name

        print(f"💾 Audio saved to: {temp_audio_path}")

        # Process with model if available
        if 'model' in globals() and 'tokenizer' in globals():
            print("🤖 Processing with Gemma 3N model...")
            response = process_colab_audio_with_model(temp_audio_path)
            print(f"\n🤖 Gemma 3N Response: {response}")

            # Optional: Convert response to speech
            if tts_method in ["pyttsx3", "gtts"]:
                speak_colab_response(response)
        else:
            print("⚠️ Model not loaded. Please run the model loading cells first.")

    except Exception as e:
        print(f"❌ Error processing Colab audio: {str(e)}")
        import traceback
        traceback.print_exc()

def process_colab_audio_with_model(audio_file_path):
    """Process the Colab audio with Gemma 3N model"""
    try:
        # Create message for the model
        messages = [{
            "role": "user",
            "content": [
                {"type": "audio", "audio": audio_file_path},
                {"type": "text", "text": "Please respond to what I said."}
            ]
        }]

        # Generate response using the model
        inputs = tokenizer.apply_chat_template(
            messages,
            add_generation_prompt=True,
            return_tensors="pt",
            tokenize=True,
            return_dict=True,
        ).to("cuda" if torch.cuda.is_available() else "cpu")

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=128,
                temperature=1.0,
                top_p=0.95,
                top_k=64,
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id
            )

        # Decode response
        response = tokenizer.decode(outputs[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)
        return response.strip()

    except Exception as e:
        print(f"❌ Model processing error: {str(e)}")
        return "I apologize, but I encountered an error processing your audio. Please try again."

def speak_colab_response(text):
    """Convert response to speech in Colab"""
    try:
        if tts_method == "gtts":
            from gtts import gTTS
            import tempfile
            import pygame

            tts = gTTS(text=text, lang='en')
            with tempfile.NamedTemporaryFile(suffix='.mp3', delete=False) as tmp_file:
                tts.save(tmp_file.name)

                # Display audio player
                from IPython.display import Audio
                display(Audio(tmp_file.name, autoplay=True))
                print("🔊 Audio response generated!")

        elif tts_method == "pyttsx3":
            print("🔊 Text-to-speech not available in Colab browser environment")
            print("📝 Response displayed as text above")
        else:
            print("🔊 Text-to-speech not available")

    except Exception as e:
        print(f"❌ TTS Error: {str(e)}")

# Register the callback for Colab
if 'google.colab' in sys.modules:
    output.register_callback('process_colab_audio', process_colab_audio)
    print("✅ Colab audio system ready!")
else:
    print("⚠️ Not in Google Colab - use local audio recording instead")
```

```python
from IPython.display import HTML, display, Audio
from gtts import gTTS
import tempfile
import os
import base64
import threading
import time
import numpy as np

class ContinuousConversation:
    """Streamlined continuous conversation system for Colab"""

    def __init__(self):
        self.conversation_history = []
        self.is_active = False

    def process_audio_input(self, audio_data):
        """Process audio input and return spoken response"""
        try:
            # Save audio temporarily with .webm extension
            with tempfile.NamedTemporaryFile(suffix='.webm', delete=False) as tmp_file:
                tmp_file.write(audio_data)
                temp_path = tmp_file.name

            # Create message for model
            messages = []

            # Add recent history for context
            if self.conversation_history:
                messages.extend(self.conversation_history[-4:])

            # Add current message
            messages.append({
                "role": "user",
                "content": [
                    {"type": "audio", "audio": temp_path},
                    {"type": "text", "text": "Please respond to what I said."}
                ]
            })

            # Get model response
            response = self.get_model_response(messages)
            print(f"\n🤖 Response: {response}\n")

            # Add to history
            self.conversation_history.append({
                "role": "assistant",
                "content": [{"type": "text", "text": response}]
            })

            # Generate and play voice response
            self.speak_response(response)

            # Cleanup
            try:
                os.unlink(temp_path)
            except:
                pass

            return response

        except Exception as e:
            print(f"❌ Error: {str(e)}")
            import traceback
            traceback.print_exc()
            return "Sorry, I couldn't process that. Please try again."

    def get_model_response(self, messages):
        """Get response from the model"""
        try:
            # Generate model inputs
            inputs = tokenizer.apply_chat_template(
                messages,
                add_generation_prompt=True,
                return_tensors="pt",
                tokenize=True,
                return_dict=True,
            ).to(model.device)

            # Generate response
            with torch.no_grad():
                outputs = model.generate(
                    **inputs,
                    max_new_tokens=128,
                    temperature=1.0,
                    top_p=0.95,
                    top_k=64,
                    do_sample=True,
                    pad_token_id=tokenizer.eos_token_id
                )

            # Decode and return response
            response = tokenizer.decode(
                outputs[0][inputs['input_ids'].shape[1]:],
                skip_special_tokens=True
            ).strip()

            return response

        except Exception as e:
            print(f"❌ Model error: {str(e)}")
            import traceback
            traceback.print_exc()
            return "I encountered an error. Please try again."

    def speak_response(self, text):
        """Convert response to speech and play it"""
        try:
            print("🔊 Generating voice response...")

            # Create temporary file for audio
            with tempfile.NamedTemporaryFile(suffix='.mp3', delete=False) as tmp_file:
                # Generate speech
                tts = gTTS(text=text, lang='en', slow=False)
                tts.save(tmp_file.name)

                # Display audio player with autoplay
                display(Audio(filename=tmp_file.name, autoplay=True))
                print("✅ Voice response ready!")

                # Schedule file cleanup
                def cleanup():
                    time.sleep(30)  # Wait for playback
                    try:
                        os.unlink(tmp_file.name)
                    except:
                        pass

                threading.Thread(target=cleanup, daemon=True).start()

        except Exception as e:
            print(f"❌ TTS error: {str(e)}")
            print("📝 Text response shown above")

    def start(self):
        """Start the conversation"""
        self.is_active = True
        print("🎙️ Started! Click the microphone button to speak.")
        create_colab_audio_recorder()

    def stop(self):
        """Stop the conversation"""
        self.is_active = False
        print("🛑 Conversation ended")
```

```python
!wget -qqq https://upload.wikimedia.org/wikipedia/en/6/61/I_Have_A_Dream_sample.ogg

# Initialize conversation system
conversation = ContinuousConversation()

# Setup audio processing callback
if 'google.colab' in sys.modules:
    def process_voice_input(audio_base64):
        """Process recorded audio"""
        if not conversation.is_active:
            print("⚠️ Please start the conversation first")
            return

        try:
            # Process audio
            audio_data = base64.b64decode(audio_base64)
            import asyncio
            response = asyncio.run(conversation.process_audio_input(audio_data))
            print(f"\n🤖 Response: {response}\n")

        except Exception as e:
            print(f"❌ Error: {str(e)}")

    # Register callback
    output.register_callback('process_colab_audio', process_voice_input)
    print("✅ Voice conversation system ready!")
```
